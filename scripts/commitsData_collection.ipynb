```python
# set key,value pair for repos
key= 'antlr'
value= 'antlr/antlr4'
```

```python
g = Github(access_token)
repo = g.get_repo(value)

#Get commits to a repo
df_rows=[]
commits = repo.get_commits()

for commit in commits:
    try:
        for file in commit.files:
            rows=[]
            rows.append(commit.sha)
            rows.append(commit.commit.message.split('\n\n')[0])
            rows.append(commit.commit.author.name)
            rows.append(commit.commit.author.date)
            rows.append(commit.stats.additions)
            rows.append(commit.stats.deletions)
            rows.append(len(commit.files))
            rows.append(file.sha)
            rows.append(file.filename)
            rows.append(file.additions)
            rows.append(file.deletions)
            df_rows.append(rows)
    except:
        continue    


# create dataframe
columns=['SHA','title','committer','date','commit additions','commit deletions','files changed',
         'file sha','filename','file additions','file deletions']
df=pd.DataFrame(df_rows,columns=columns)
df['file changes']=df['file additions'] + df['file deletions']
# save dataframe
df.to_csv('data/commits/'+key+'.csv')
```
